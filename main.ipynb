```python
import pandas as pd
```

```python
adjectives = pd.read_json('data/adjectives.json')
adjectives[:3]
```

```python
nouns = pd.read_json('data/nouns.json')
nouns[:3]
```

```python
from detoxify import Detoxify
model = Detoxify('original', device='cuda')
```

```python
adj_res = model.predict(adjectives[0].tolist())
```

```python
noun_res = model.predict(nouns[0].tolist())
```

```python
adjectives = pd.DataFrame(adj_res, index=adjectives[0]).round(5)
adjectives[:3]
```

```python
nouns = pd.DataFrame(noun_res, index=nouns[0]).round(5)
nouns[:3]
```

```python
profanity_adjectives = adjectives[adjectives["toxicity"] > 0.5]
profanity_adjectives.sort_values("toxicity", ascending=False)
profanity_adjectives["toxicity"].to_json("data/profanity_adjectives.json", indent=2)
profanity_adjectives
```

```python
profanity_nouns = nouns[nouns["toxicity"] > 0.5]
profanity_nouns.sort_values("toxicity", ascending=False)
profanity_nouns["toxicity"].to_json("data/profanity_nouns.json", indent=2)
profanity_nouns
```
